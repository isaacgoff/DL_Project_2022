```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
```

```python
#select music file, custom path required
music_file ='nsynth-test/audio/guitar_electronic_022-038-050.wav' 
```

```python
#load the waveform y and sampling rate sr
y, sr = librosa.load(music_file, sr=None)
```

```python
print("librosa gives us an array of type: " + str(type(y)))
print("and of shape: " + str(y.shape))
print("sample rate is: " + str(sr) + " samples a second")
```

```python
#print in time domain
plt.plot(y)
```

```python
#convert to 2 dimensional spectogram format
spectogram = librosa.feature.melspectrogram(y = y, sr = sr, n_mels=128, fmax=8000, hop_length = 502 )
```

```python
print("librosa gives us an array of type: " + str(type(spectogram)))
print("and of shape: " + str(spectogram.shape))
```

```python
#plot spectogram:
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(spectogram, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
```
